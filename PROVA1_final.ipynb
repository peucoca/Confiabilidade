```python
def sobrevivencia_serie(eq,t):

  Rs = np.exp(-eq*t) # Probabilidade de falha 
  Qs = 1 - Rs # Probabilidade de sobrevivncia
  

  return Rs,Qs
```

```python
def sobrevivencia_paralelo(,t,n): # Sistema paralelo no possui taxa de falha equivalente

  aux = [0 for i in range(0, len())]

  for i in range(len()):
    
    aux[i] = (1 - np.exp(-[i]*t))**n[i]  # Cojunto de
     
     
  aux2 = np.prod(aux)
  Rs = 1 - aux2 # Probabilidade de falha 
  Qs = aux2 # Probabilidade de sobrevivncia

  print("Probabilidade de falha em",t,'horas :',Rs)
  print('                                              ')
  print("Probabilidade de sobrevivncia em",t,'horas :',Qs)

  return Rs,Qs
```

```python
def factorial(n):
 
    f = 1
    for i in range(2, n+1):
        f *= i
         
    return f

# Funo para imprimir a srie
def series(A, X, n):

    #Calculando o valor de n!
    nFact = factorial(n)
    valor = []
    #Loop para mostrar a srie
  
    for i in range(0, n + 1):
        #Calcula o valor de nCr
        niFact = factorial(n - i)
        iFact = factorial(i)

        #Calculando o valor de A^k e X^K
        aPow = pow(A, n - i)
        xPow = pow(X, i)
 
        # Funo para imprimir a srie
        aux = int((nFact * aPow * xPow) / (niFact * iFact))
        #print (int((nFact * aPow * xPow) /
                   #(niFact * iFact)), end = " ")

        valor.append(aux) # Criei uma lista para armazenar os valores dos coeficientes
        
    return valor # Retorna os coefiecientes da expanso
     


```
